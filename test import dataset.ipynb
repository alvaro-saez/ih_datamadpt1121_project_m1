#### obtener latitud y longitud de bicimad

```python
def bicimad_geo_long(geo):
    geo_clean = geo.replace("[","").replace("]","").split(",")
    longitud = geo_clean[0].strip()
    return longitud
```

```python
bicimad_st_df["LONGITUD"] = bicimad_st_df["geometry_coordinates"].apply(bicimad_geo_long)
bicimad_st_df.head()
```

```python
def bicimad_geo_lat(geo):
    geo_clean = geo.replace("[","").replace("]","").split(",")
    latitud = geo_clean[1].strip()
    return latitud
```

```python
bicimad_st_df["LATITUD"] = bicimad_st_df["geometry_coordinates"].apply(bicimad_geo_lat)
bicimad_st_df.head()
```

```python
final_df["Station latitude"] = pd.to_numeric(final_df["Station latitude"])
final_df["Station longitude"] = pd.to_numeric(final_df["Station longitude"])
final_df.info()
```

```python
def to_mercator(lat, long):
    # transform latitude/longitude data in degrees to pseudo-mercator coordinates in metres
    c = gpd.GeoSeries([Point(lat, long)], crs=4326)
    c = c.to_crs(3857)
    return c
```

```python
def distance_meters(lat_start, long_start, lat_finish, long_finish):
    # return the distance in metres between to latitude/longitude pair point in degrees (i.e.: 40.392436 / -3.6994487)
    start = to_mercator(lat_start, long_start)
    finish = to_mercator(lat_finish, long_finish)
    return start.distance(finish)
```

```python
final_df["Distance Between BiciMAD station and Place of interest"] = final_df.apply(lambda final_df: distance_meters(final_df["Place latitude"], final_df["Place longitude"], final_df["Station latitude"], final_df["Station longitude"]), axis=1)
```
